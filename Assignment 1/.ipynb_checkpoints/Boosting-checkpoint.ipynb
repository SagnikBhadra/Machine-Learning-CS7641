```python
def define_Boosting_Classifier(input_train_set, output_train_set):
    AdaBoost_Learner = AdaBoostClassifier()
    AdaBoost_Learner.fit(input_train_set, output_train_set.values.ravel())
    
    print(AdaBoost_Learner)
    return AdaBoost_Learner

def boosting_predictions(AdaBoost_Learner, input_train_set, output_train_set):
    # first net results
    y_hat = AdaBoost_Learner.predict(input_train_set)
    print(classification_report(output_train_set, y_hat))


def compute_score(input_train_set, output_train_set, AdaBoost_Learner, mean):
    result = stratFold(input_train_set, output_train_set, AdaBoost_Learner, f1_score, average=mean)
    plot_learning_curve(result[0], result[1], result[2], title="f1_macro score learning curve")
    

def run_gridsearch(values, adaBoost, input_train_set, output_train_set, cross_validation_splitter, weighting, parallel):
    grid_suggestion = GridSearchCV(adaBoost, values, cv=cross_validation_splitter, scoring=weighting, n_jobs = parallel)
    grid_suggestion.fit(input_train_set, output_train_set.values.ravel())
    print(grid_suggestion.best_estimator_, grid_suggestion.best_score_)
    
    return grid_suggestion
    
    
def run_validation_curve(AdaBoost_Learner, input_train_set, output_train_set, hyperparameter_values, weighting, hyperparameter, inverse_x, cross_validation_splitter, title):
    result_train, result_test = validation_curve(
        AdaBoost_Learner, input_train_set, output_train_set.values.ravel(), param_name=hyperparameter, param_range=hyperparameter_values,
        scoring=weighting, verbose=1, cv=cross_validation_splitter, n_jobs=-1
    )
    print(result_train.mean(axis=1), result_test.mean(axis=1), hyperparameter_values)
    plot_complexity(result_train.mean(axis=1), result_test.mean(axis=1), hyperparameter_values, 
                    title='f1_weighted score', inverse_x=inverse_x)
    
    return result_train, result_test

def wine_dataset():
    target_feature = "class"
    mean = "weighted"
    weighting = "accuracy"
    hyperparameter = "n_estimators"
    cross_validation_splitter = 10
    parallel = -1
    inverse_x = False
    
    
    np.random.seed(44)
    wine_dataset = pd.read_csv("wine_dataset.csv")
    print(wine_dataset.info())
    wine_dataset.head()
    
    target = wine_dataset["class"]
    wine_dataset = normalize_data(wine_dataset)
    wine_dataset["class"] = target
    
    in_sample, out_of_sample = create_stratified_data(target_feature, wine_dataset)
    
    output_train_set, input_train_set, output_test_set, input_test_set = prepare_data(in_sample, out_of_sample, target_feature)
    
    AdaBoost_Learner = define_Boosting_Classifier(input_train_set, output_train_set)
    
    boosting_predictions(AdaBoost_Learner, input_train_set, output_train_set)
    
    result = cross_val_score(AdaBoost_Learner, input_train_set, output_train_set.values.ravel(), scoring=weighting, cv=cross_validation_splitter)
    print(result.mean())
    
    AdaBoost_Learner = AdaBoostClassifier()
    compute_score(input_train_set, output_train_set, AdaBoost_Learner, mean)
    
    
    values = {
    'base_estimator__max_depth': [1, 2, 3, None],
    "base_estimator__splitter" : ["best", "random"],
    'n_estimators': [25, 50, 100],
    'base_estimator__criterion': ['gini', 'entropy']
    }
    DT_Learner = DecisionTreeClassifier(max_depth=None)
    adaBoost = AdaBoostClassifier(DT_Learner)
    weighting='f1_weighted'
    grid_suggestion = run_gridsearch(values, adaBoost, input_train_set, output_train_set, cross_validation_splitter, weighting, parallel)

    print(grid_suggestion.best_estimator_, grid_suggestion.best_score_)

    compute_score(input_train_set, output_train_set, grid_suggestion.best_estimator_, mean)
    
    values = {
        'base_estimator__max_depth': [1, 2],
        "base_estimator__splitter" : ["best"],
        'n_estimators': [30, 50, 60, 70,],
        'base_estimator__criterion': ['gini']
    }
    DT_Learner = DecisionTreeClassifier(max_depth=None)
    adaBoost = AdaBoostClassifier(DT_Learner)
    
    
    weighting='f1_weighted'
    grid_suggestion = run_gridsearch(values, adaBoost, input_train_set, output_train_set, cross_validation_splitter, weighting, parallel)

    print(grid_suggestion.best_estimator_, grid_suggestion.best_score_)

    compute_score(input_train_set, output_train_set, grid_suggestion.best_estimator_, mean)
    
    
    values = {
        'base_estimator__max_depth': [1, 2],
        "base_estimator__splitter" : ["best"],
        'n_estimators': [50, 60, 70, 80],
        'learning_rate': [0.5, 0.8, 1.0, 1.2]
    }
    DT_Learner = DecisionTreeClassifier(max_depth=None)
    adaBoost = AdaBoostClassifier(DT_Learner)
    
    grid_suggestion = run_gridsearch(values, adaBoost, input_train_set, output_train_set, cross_validation_splitter, weighting, parallel)

    print(grid_suggestion.best_estimator_, grid_suggestion.best_score_)

    compute_score(input_train_set, output_train_set, grid_suggestion.best_estimator_, mean)
    
    
    boosted_tree = AdaBoostClassifier(algorithm='SAMME.R',
                   base_estimator=DecisionTreeClassifier(ccp_alpha=0.0,
                                                         class_weight=None,
                                                         criterion='gini',
                                                         max_depth=1,
                                                         max_features=None,
                                                         max_leaf_nodes=None,
                                                         min_impurity_decrease=0.0,
                                                         min_impurity_split=None,
                                                         min_samples_leaf=1,
                                                         min_samples_split=2,
                                                         min_weight_fraction_leaf=0.0,
                                                         random_state=None,
                                                         splitter='best'),
                   learning_rate=0.5, n_estimators=70, random_state=None)
    
    hyperparameter_values = np.linspace(10, 150, 10).astype(int)
    title='ROC Complexity Curve on n_estimators'
    result_train, result_test = run_validation_curve(boosted_tree, input_train_set, output_train_set, hyperparameter_values, weighting, hyperparameter, inverse_x, cross_validation_splitter, title)
    
    optimal_Boosting_Learner = AdaBoostClassifier(algorithm='SAMME.R',
                   base_estimator=DecisionTreeClassifier(ccp_alpha=0.0,
                                                         class_weight=None,
                                                         criterion='gini',
                                                         max_depth=1,
                                                         max_features=None,
                                                         max_leaf_nodes=None,
                                                         min_impurity_decrease=0.0,
                                                         min_impurity_split=None,
                                                         min_samples_leaf=1,
                                                         min_samples_split=2,
                                                         min_weight_fraction_leaf=0.0,
                                                         random_state=None,
                                                         splitter='best'),
                   learning_rate=0.5, n_estimators=110, random_state=None)
    compute_score(input_train_set, output_train_set, optimal_Boosting_Learner, mean)
    
    optimal_Boosting_Learner.fit(input_train_set, output_train_set.values.ravel())
    test_y_hat = optimal_Boosting_Learner.predict(input_test_set)
    print(classification_report(output_test_set, test_y_hat, digits=5))
    print(confusion_matrix(output_test_set, test_y_hat))


def ufc_dataset():
    target_feature = "Winner"
    alpha = 0.005
    criterion = "entropy"
    mean = "macro"
    weighting = "f1_macro"
    hyperparameter = "learning_rate"
    cross_validation_splitter = 10
    parallel = -1
    inverse_x = False
    
    
    np.random.seed(44)
    ufc_dataset = pd.read_csv("ufc_data.csv")
    print(ufc_dataset.info())
    ufc_dataset.head()
    
    # red is 1, blue is 0
    # drop the no_of_rounds because unknown in the beginning
    ufc_dataset.loc[ufc_dataset["Winner"] == "Red", "Winner"] = 1
    ufc_dataset.loc[ufc_dataset["Winner"] == "Blue", "Winner"] = 0
    ufc_dataset.drop(["no_of_rounds"], axis=1 ,inplace=True)
    # al zero
    ufc_dataset.drop(["B_draw"], axis=1 ,inplace=True)
    ufc_dataset.drop(["R_draw"], axis=1 ,inplace=True)
    
    ufc_dataset = normalize_data(ufc_dataset)
    
    in_sample, out_of_sample = create_stratified_data(target_feature, ufc_dataset)
    
    train_set = in_sample
    test_set = out_of_sample
    
    output_train_set, input_train_set, output_test_set, input_test_set = prepare_data(train_set, test_set, target_feature)
    
    AdaBoost_Learner = define_Boosting_Classifier(input_train_set, output_train_set)
    
    boosting_predictions(AdaBoost_Learner, input_train_set, output_train_set)
    
    scores = cross_val_score(AdaBoost_Learner, input_train_set, output_train_set.values.ravel(), scoring=weighting, cv=cross_validation_splitter)
    scores.mean()
    
    AdaBoost_Learner = AdaBoostClassifier()
    compute_score(input_train_set, output_train_set, AdaBoost_Learner, mean)
    
    
    values = {
        'base_estimator__max_depth': [1, 2],
        "base_estimator__splitter" : ["best", "random"],
        'n_estimators': [25, 50, 75]
    }
    DT_Learner = DecisionTreeClassifier(max_depth=None, class_weight="balanced")
    adaBoost = AdaBoostClassifier(DT_Learner)
    
    grid_suggestion = run_gridsearch(values, adaBoost, input_train_set, output_train_set, cross_validation_splitter, weighting, parallel)

    print(grid_suggestion.best_estimator_, grid_suggestion.best_score_)

    compute_score(input_train_set, output_train_set, grid_suggestion.best_estimator_, mean)
    
    values = {
        "base_estimator__max_depth": [1 , 2, 3],
        "base_estimator__splitter" : ["best", "random"],
        'n_estimators': [50, 60, 70, 80, 90],
        'base_estimator__criterion': ['entropy', 'gini']
    }
    DT_Learner = DecisionTreeClassifier(class_weight='balanced')
    adaBoost = AdaBoostClassifier(DT_Learner)
    
    
    grid_suggestion = run_gridsearch(values, adaBoost, input_train_set, output_train_set, cross_validation_splitter, weighting, parallel)

    print(grid_suggestion.best_estimator_, grid_suggestion.best_score_)

    compute_score(input_train_set, output_train_set, grid_suggestion.best_estimator_, mean)
    
    
    DT_Learner = DecisionTreeClassifier(max_depth=1, splitter="random", 
                                             class_weight='balanced')
    adaBoost = AdaBoostClassifier(DT_Learner, learning_rate=1.0, 
                                  n_estimators=70)
    hyperparameter_values = np.linspace(0.2,1.2,22)
    title='F1 Macro Complexity Curve on learning_rate'
    result_train, result_test = run_validation_curve(adaBoost, input_train_set, output_train_set, hyperparameter_values, weighting, hyperparameter, inverse_x, cross_validation_splitter, title)
    
    DT_Learner = DecisionTreeClassifier(max_depth=1, splitter="random", 
                                                 class_weight='balanced')
    adaBoost = AdaBoostClassifier(DT_Learner, learning_rate=0.55, 
                                  n_estimators=70)
    hyperparameter_values = np.linspace(10, 150, 10).astype(int)
    hyperparameter="n_estimators"
    title='F1 Macro Complexity Curve on n_estimators'
    result_train, result_test = run_validation_curve(adaBoost, input_train_set, output_train_set, hyperparameter_values, weighting, hyperparameter, inverse_x, cross_validation_splitter, title)
    
    optimal_Boosting_Learner = AdaBoostClassifier(algorithm='SAMME.R',
                   base_estimator=DecisionTreeClassifier(ccp_alpha=0.0,
                                                         class_weight='balanced',
                                                         criterion='gini',
                                                         max_depth=1,
                                                         max_features=None,
                                                         max_leaf_nodes=None,
                                                         min_impurity_decrease=0.0,
                                                         min_impurity_split=None,
                                                         min_samples_leaf=1,
                                                         min_samples_split=2,
                                                         min_weight_fraction_leaf=0.0,
                                                         random_state=None,
                                                         splitter='random'),
                   learning_rate=0.55, n_estimators=100, random_state=None)
    
    
    compute_score(input_train_set, output_train_set, optimal_Boosting_Learner, mean)

    
    optimal_Boosting_Learner.fit(input_train_set, output_train_set.values.ravel())
    test_y_hat = optimal_Boosting_Learner.predict(input_test_set)
    print(classification_report(output_test_set, test_y_hat, digits=5))
    print(confusion_matrix(output_test_set, test_y_hat))
    
    
if __name__ == "__main__":
    wine_dataset()
    ufc_dataset()
```
